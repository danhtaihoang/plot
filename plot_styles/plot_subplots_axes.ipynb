```python
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
%matplotlib inline
```

```python
np.random.seed(1)
```

```python
bins = np.linspace(0.,1.,11,endpoint=True)
bins2 = bins[:-1] + 0.05
```

```python
p11 = np.loadtxt('hist_g2_size02.txt')
p12 = np.loadtxt('hist_g2_size08.txt')
p21 = np.loadtxt('hist_g10_size02.txt')
p22 = np.loadtxt('hist_g10_size08.txt')
```

```python
acc1 = np.loadtxt('acc_g2.txt')
acc2 = np.loadtxt('acc_g10.txt')
```

```python
list_train_size = [0.8,0.6,0.4,0.2]
```

```python
#plt.figure(figsize=(11,6))
fig, axes = plt.subplots(2, 3,figsize=(11,6))

axes[0,0].text(-0.18,0.33,'(a)',fontsize=13)
axes[0,0].bar(bins2,p11[0],width=0.095,label='class A')
axes[0,0].bar(bins2,p11[1],width=0.095,alpha=0.5,label='class B')
axes[0,0].legend()

axes[0,1].text(-0.18,0.33,'(b)',fontsize=13)
axes[0,1].bar(bins2,p12[0],width=0.095)
axes[0,1].bar(bins2,p12[1],width=0.095,alpha=0.5)

axes[0,2].text(-0.02,0.792,'(c)',fontsize=13)
axes[0,2].plot(list_train_size,acc1[:,0],'k--',marker='o',mfc='none',label='Logistic Regression')
axes[0,2].plot(list_train_size,acc1[:,1],'b--',marker='s',mfc='none',label='Naive Bayes')
axes[0,2].plot(list_train_size,acc1[:,2],'r--',marker='^',mfc='none',label='Random Forest')
axes[0,2].plot(list_train_size,acc1[:,-1],'k-',marker='o',label='Expectation Reflection')
axes[0,2].set_yticks([0.65,0.7,0.75,0.8])
axes[0,2].legend()

axes[1,0].text(-0.18,0.33,'(d)',fontsize=13)
axes[1,0].bar(bins2,p21[0],width=0.095)
axes[1,0].bar(bins2,p21[1],width=0.095,alpha=0.5)

axes[1,1].text(-0.18,0.33,'(e)',fontsize=13)
axes[1,1].bar(bins2,p22[0],width=0.095)
axes[1,1].bar(bins2,p22[1],width=0.095,alpha=0.5)

axes[1,2].text(0.05,0.926,'(f)',fontsize=13)
axes[1,2].plot(list_train_size,acc2[:,0],'k--',marker='o',mfc='none',label='Logistic Regression')
axes[1,2].plot(list_train_size,acc2[:,1],'b--',marker='s',mfc='none',label='Naive Bayes')
axes[1,2].plot(list_train_size,acc2[:,2],'r--',marker='^',mfc='none',label='Random Forest')
axes[1,2].plot(list_train_size,acc2[:,-1],'k-',marker='o',label='Expectation Reflection')
axes[1,2].set_yticks([0.75,0.8,0.85,0.9])

for i in range(2):
    for j in range(2):
        axes[i,j].plot([0.5,0.5],[0.0,0.35],'b--')
        axes[i,j].set_xlim([0,1])
        axes[i,j].set_ylim([0,0.35])
        axes[i,j].set_xlabel('Probability')
        axes[i,j].set_ylabel('Historgam')
        axes[i,j].set_xticks([0,0.5,1])
        axes[i,j].set_yticks([0,0.1,0.2,0.3])

for i in range(2):
    axes[i,2].set_xlabel('training size')
    axes[i,2].set_ylabel('accuracy')
    axes[i,2].yaxis.set_minor_locator(AutoMinorLocator(5))
    axes[i,2].set_xticks([0.2,0.4,0.6,0.8],('20%','40%','60%','80%'))
        
plt.tight_layout(h_pad=1, w_pad=1.5)
plt.savefig('fig1.pdf', format='pdf', dpi=100)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
gs = gridspec.GridSpec(3, 3)
ax1 = plt.subplot(gs[0, :])
ax2 = plt.subplot(gs[1,:-1])
ax3 = plt.subplot(gs[1:, -1])
ax4 = plt.subplot(gs[-1,0])
ax5 = plt.subplot(gs[-1,-2])

ax1.set_title('ax1')
ax2.set_title('ax2')
ax3.set_title('ax3')
ax4.set_title('ax4')
ax5.set_title('ax5')

plt.tight_layout(h_pad=1, w_pad=1.5)
```

```python
## Adjust GridSpec layout
nx,ny = 2,3
fig = plt.figure(figsize=(nx*4.5,ny*2.5))

gs1 = gridspec.GridSpec(3, 2)
gs1.update(left=0.08, right=0.48, wspace=0.5, hspace=0.5)
ax1 = plt.subplot(gs1[0, :2])
ax2 = plt.subplot(gs1[1, 0])
ax3 = plt.subplot(gs1[1, 1])
ax4 = plt.subplot(gs1[2, 0])

gs2 = gridspec.GridSpec(3, 1)
gs2.update(left=0.6, right=0.98, wspace=0.5,hspace=0.5)
ax5 = plt.subplot(gs2[0, 0])
ax6 = plt.subplot(gs2[1, 0])
ax7 = plt.subplot(gs2[2, 0])

ax1.set_title('ax1')
ax2.set_title('ax2')
ax3.set_title('ax3')
ax4.set_title('ax4')
ax5.set_title('ax5')
ax6.set_title('ax6')
ax7.set_title('ax7')

#plt.tight_layout(h_pad=1, w_pad=1.5)
#plt.savefig('fig_gs.pdf', format='pdf', dpi=100)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os,sys,time
```

```python
projects = ["KIRC", "KIRP", "KICH", "LUSC", "LUAD", "BRCA", "LGG",
            "PRAD", "HNSC", "COAD", "STAD", "READ", "CESC", 
            "ESCA", "PAAD", "GBM"]

print(len(projects))

analysis_results = "analysis_results/"
```

```python
n_sign_genes_all = np.zeros(len(projects))
for ip,project in enumerate(projects):
    df_padj = pd.read_csv(f"{analysis_results}/{project}_R.csv")

    p_adj = df_padj["p_adj"].values
    n_sign_genes = np.sum(p_adj < 0.05)
    print(n_sign_genes)

    n_sign_genes_all[ip] = n_sign_genes
```

```python
df = pd.DataFrame(data = {"cohort": projects, "n_sign_genes": n_sign_genes_all})
df.to_csv(f"{analysis_results}n_sign_genes_16TCGA.csv", index=None)

df_sorted = df.sort_values(by="n_sign_genes", ascending=False)
df_sorted.to_csv(f"{analysis_results}n_sign_genes_16TCGA_sorted.csv", index=None)
```
